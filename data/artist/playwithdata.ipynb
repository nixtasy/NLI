```python
import pandas as pd
```

```python
with open('songs_train.txt', 'r') as f:
    train = f.readlines()

with open('songs_dev.txt', 'r') as f:
    dev = f.readlines()

with open('songs_test.txt', 'r') as f:
    test = f.readlines()
```

```python
train_artists = []
train_titles = []
train_lyrics = []
for item in train:
    artist, title, lyric = item.split("\t")
    train_artists.append(artist)
    train_titles.append(title)
    train_lyrics.append(lyric.strip())
```

```python
zippedList =  list(zip(train_artists, train_titles, train_lyrics))
train_df = pd.DataFrame(zippedList, columns = ['artist' , 'title', 'lyrics'],) 
```
